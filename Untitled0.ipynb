```python
# YOLOv8 설치
!pip install ultralytics

# 필요한 라이브러리 불러오기
from ultralytics import YOLO
import matplotlib.pyplot as plt
import cv2
import numpy as np
from google.colab import files
from PIL import Image, ImageDraw, ImageFont
import io

# 이미지 업로드
uploaded = files.upload()

# 업로드한 파일명 확인
image_path = list(uploaded.keys())[0]

# 더 성능이 좋은 YOLOv8 모델 로드 (중간 크기 또는 큰 모델 사용)
model = YOLO("yolov8m.pt")  # m은 medium 모델, x는 가장 큰 모델

# 신뢰도 임계값을 낮춰서 더 많은 객체 검출 (기본값은 0.25)
results = model(image_path, conf=0.2)

# 고양이 클래스(클래스 번호 15) 필터링
cats = [box for box in results[0].boxes if int(box.cls) == 15]
cat_count = len(cats)

# 원본 이미지 로드
img = Image.open(io.BytesIO(uploaded[image_path]))
img_np = np.array(img)

# 결과 시각화 - 고양이만 강조 표시
for cat in cats:
    # 박스 좌표
    x1, y1, x2, y2 = map(int, cat.xyxy[0])
    # 신뢰도 점수
    conf = float(cat.conf[0])

    # 원본 이미지에 고양이 표시 (빨간색 박스로 강조)
    cv2.rectangle(img_np, (x1, y1), (x2, y2), (255, 0, 0), 2)
    cv2.putText(img_np, f"Cat: {conf:.2f}", (x1, y1-10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

# 결과 출력
print(f'고양이 수: {cat_count}')
print(f'신뢰도 임계값: 0.2 (기본값 0.25보다 낮춤)')

if cat_count == 0:
    print("고양이가 감지되지 않았습니다. 다음을 시도해보세요:")
    print("1. 더 큰 YOLOv8 모델(yolov8l.pt 또는 yolov8x.pt) 사용")
    print("2. 신뢰도 임계값을 더 낮추기 (conf=0.1 시도)")

# 원본 이미지와 처리된 이미지 함께 표시
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(Image.open(io.BytesIO(uploaded[image_path])))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title(f"Detected Cats: {cat_count}")
plt.imshow(img_np)
plt.axis('off')

plt.tight_layout()
plt.show()

# 전체 객체 탐지 결과도 확인
plt.figure(figsize=(10, 8))
plt.title("All Objects Detected")
plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
```
