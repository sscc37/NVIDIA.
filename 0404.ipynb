```python
# YOLOv8 설치
!pip install ultralytics

# 필요한 라이브러리 불러오기
from ultralytics import YOLO
import matplotlib.pyplot as plt
from google.colab import files

# 이미지 업로드
uploaded = files.upload()

# 업로드한 파일명 확인
image_path = list(uploaded.keys())[0]

# YOLOv8 모델 로드 (사전 학습된 모델 사용)
model = YOLO("yolov8n.pt")  # yolov8n.pt는 YOLOv8의 작은 모델입니다. 다른 모델도 가능

# 이미지에서 객체 탐지 수행
results = model(image_path)

# 결과에서 '사람' 클래스(클래스 번호 0)만 필터링
# YOLOv8에서 사람은 클래스 0에 해당합니다.
people_count = sum([1 for cls in results[0].boxes.cls if cls == 0])

# 결과 출력
print(f'사람 수: {people_count}')

# 탐지 결과 이미지 표시
plt.imshow(results[0].plot())
plt.axis('off')
plt.show()
```
